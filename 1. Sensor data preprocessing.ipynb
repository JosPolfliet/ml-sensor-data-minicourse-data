# Plotting and pre-processing sensor data

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv('https://raw.githubusercontent.com/JosPolfliet/ml-sensor-data-minicourse-data/master/accelerometer_500k.csv', names=['time', 'x', 'y','z']).iloc[10000:20000]
df['time'] = pd.to_datetime(df['time'], unit='s')
df = df.set_index('time')
df
```

## Smoothing - basics
Try experimenting with the *rolling windowsize* parameter to see how it affects the output.

More kernels are availabale in the `win_types` parameter (see https://en.wikipedia.org/wiki/Window_function)

```python
df['sm_x'] = df['x'].rolling(window=500, center=True).mean()
df['x'].plot(figsize=(20,16), linewidth=0.5)
df['sm_x'].plot(figsize=(20,16), linewidth=2)
```

## Simple anomaly detection based on moving average

```python
df['x_rolling_mean'] = df['x'].rolling(1000, center=True).mean()

df['x_ul'] = df['x_rolling_mean'] + 3 * df['x'].std()
df['x_ll'] = df['x_rolling_mean'] - 3 * df['x'].std()
df['x'].plot(figsize=(20,16), linewidth=0.5)
for col in [ 'x_rolling_mean', 'x_ul', 'x_ll']:
    df[col].plot(figsize=(20,16))
```

## Slightly less simple anomaly detecttion based on moving average with moving standard deviation bounds

```python
df['x_rolling_mean'] = df['x'].rolling(1000, center=True).mean()
df['x_rolling_std'] = df['x'].rolling(1000).std()
df['x_ul'] = df['x_rolling_mean'] + 3 * df['x_rolling_std'] 
df['x_ll'] = df['x_rolling_mean'] - 3 * df['x_rolling_std'] 
df['x'].plot(figsize=(20,16), linewidth=0.5)
for col in [ 'x_rolling_mean', 'x_ul', 'x_ll']:
    df[col].plot(figsize=(20,16))

plt.show()
df['x_rolling_std'].plot(figsize=(20,5), subplots=True)
```
